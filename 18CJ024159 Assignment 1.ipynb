```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

raw_data = pd.read_csv('data.csv')

for i in range(2010, 2020, 1):
    data = raw_data[:54][['Country Name', str(i)]]
    location = data.iloc[:,0]
    year = data.iloc[:,1]
    index = np.arange(len(data))
    plt.figure(figsize=(20,5))

    plt.bar(location, year, color='red', width = 1, edgecolor = 'black', align = 'center')

    plt.xlabel("Countries", fontsize = 18)
    plt.ylabel("% Government expenditure", fontsize = 18)

    plt.title("Government expenditure on education as % total government expenditure in " + str(i), fontsize = 20)
    plt.xticks(index, location, fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.xticks(rotation=90)
    plt.show()
    
    std = np.std(year)
    print('Standard deviation for year ' +str(i) + ' is ' + str(std))
    vari = np.var(year)
    print('Variance for year ' +str(i) + ' is ' + str(vari))
    rangec = np.max(year) - np.min(year)
    print('Range for year ' +str(i) + ' is ' + str(rangec))
    q75, q25 = np.percentile(year, [75 ,25])
    qint = q75-q25
    print('Interquartile range for year ' +str(i) + ' is ' + str(qint))
    
```
